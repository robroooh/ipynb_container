```python
%matplotlib inline
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
from scipy import ndimage
import cmath
import math
```

```python
xs = np.linspace(0,2*np.pi*20,100)
y = [math.sin(x*2) + 1.5*math.sin(x) + 1.5*math.sin(x) for x in xs]
```

```python
plt.plot(xs,y)
```

```python
y_fft = np.fft.fft(y)
plt.plot(xs,y_fft)
```
