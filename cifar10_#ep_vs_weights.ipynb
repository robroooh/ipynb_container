```
from collections import Counter
import numpy as np


class ReportGenerator():

    '''
        ReportGenerator is intended to be a class where it accepts the weights and do the mathematic calculation 
    '''
    
    def __init__(self,name,w):
        if not isinstance(w,list):
            self.w = w.flatten()
        else:
            self.w = w
        self.name = name
    
    def get_omode(self):
        # Preprocessing by turning lists of floating point number into cute set of strings
        data = [('%.e' % abs(i))[1:] for i in self.w]
        data = Counter(data)
        self.omode = data.most_common(len(data))

    def get_min(self):
        return min(self.w)
    
    def get_max(self):
        return max(self.w)
    
    def gen_report(self):
        self.get_omode()
        print("Report for ", self.name)
        print(' ========================== ')
        print('the smallest weight = ', self.get_min())
        print('the biggest weight = ', self.get_max())
        print('the most mainstream order is as follow')
        print(self.omode)
        print(' ========================== ')

        
pf = open('allw_reg0ep10','rb')
all_w = pickle.load(pf)
print("Report for ", fname)
r = ReportGenerator(fname, all_w)
r.gen_report()
r.w[:10]
```
